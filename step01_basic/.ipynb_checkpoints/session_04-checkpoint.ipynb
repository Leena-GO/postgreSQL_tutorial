    + 구매링크: 

## 자료 조회

- SELECT와 연관된 기본문법을 활용하여 조회를 하도록 합니다. 


| 명령어  | 설명 |
|---|---|
| LIMIT  | 반환하는 로우의 개수를 지정한다  |
| OFFSET  | 반환하는 로우의 시작지점을 지정한다  |
| ORDER BY  | 반환하는 로우의 정렬할 때 지정한다  |
| WHERE  | 지정한 로우만 조회가 되도록 필터 기능을 한다  |

- 먼저 psql에 접속을 시작합니다. 
```bash
$ psql -U postgres
postgres 사용자의 암호:
psql (13.3)
도움말을 보려면 "help"를 입력하십시오.
```

- SELECT 문을 통해 전체 데이터를 조회합니다. 
```SQL
postgres=# SELECT * FROM developers;
 book_id |    date    |         name
---------+------------+----------------------
       1 | 2019-12-17 | "자바"
       2 | 2019-12-17 | '자바'
       3 | 2019-12-17 | Python's Programming
       4 | 2019-12-18 | 맛있는 MongoDB
       5 | 2019-12-18 | Python
       6 | 2019-12-18 | R
       7 | 2019-12-19 | Ruby
       8 | 2019-12-19 | Go's Programming
       9 | 2019-12-19 | Statistics
(9개 행)
```

- 이번에는 LIMIT 명령어를 활용합니다. 
```SQL
postgres=# SELECT * FROM developers LIMIT 5;
 book_id |    date    |         name
---------+------------+----------------------
       1 | 2019-12-17 | "자바"
       2 | 2019-12-17 | '자바'
       3 | 2019-12-17 | Python's Programming
       4 | 2019-12-18 | 맛있는 MongoDB
       5 | 2019-12-18 | Python
(5개 행)
```


- 이번에는 OFFSET 명령어를 추가해봅니다. 
- OFFSET은 프로그래밍에서 흔히 말하는 인덱스와 동일하다고 생각하면 된다. 
    + 0번째부터 시작을 한다. 
- 아래 코드는 즉, 1번째 인덱스 부터 시작하여 
```SQL
postgres=# SELECT * FROM developers LIMIT 6 OFFSET 1;
 book_id |    date    |         name
---------+------------+----------------------
       2 | 2019-12-17 | '자바'
       3 | 2019-12-17 | Python's Programming
       4 | 2019-12-18 | 맛있는 MongoDB
       5 | 2019-12-18 | Python
       6 | 2019-12-18 | R
       7 | 2019-12-19 | Ruby
(6개 행)
```

- 이번에는 name 기준으로 ORDER BY 명령어를 입력합니다. 
- 오름차순은 ASC, 내림차순은 DESC를 입력합니다. 

```SQL
postgres=# SELECT * FROM developers ORDER BY name ASC;
 book_id |    date    |         name
---------+------------+----------------------
       2 | 2019-12-17 | '자바'
       1 | 2019-12-17 | "자바"
       8 | 2019-12-19 | Go's Programming
       5 | 2019-12-18 | Python
       3 | 2019-12-17 | Python's Programming
       6 | 2019-12-18 | R
       7 | 2019-12-19 | Ruby
       9 | 2019-12-19 | Statistics
       4 | 2019-12-18 | 맛있는 MongoDB
```

- 이번에는 ORDER BY 각각 컬럼명을 순차적으로 입력하도록 합니다. 
- 우선, date로 
```SQL
postgres=# SELECT * FROM developers ORDER BY date, name;
 book_id |    date    |         name
---------+------------+----------------------
       2 | 2019-12-17 | '자바'
       1 | 2019-12-17 | "자바"
       3 | 2019-12-17 | Python's Programming
       5 | 2019-12-18 | Python
       6 | 2019-12-18 | R
       4 | 2019-12-18 | 맛있는 MongoDB
       8 | 2019-12-19 | Go's Programming
       7 | 2019-12-19 | Ruby
       9 | 2019-12-19 | Statistics
```

- WHERE 명령어를 활용하도록 합니다. SQL 비교 연산자는 다음과 같습니다. 
    + 참조: 모두를 위한 PostgreSQL
    
| 연산자  | 설명 |
|---|---|
| =  | 서로 같다  |
| <>  | 서로 다르다  |
| >  | 오른쪽 값보다 왼쪽 값이 더 크다  |
| <  | 왼쪽 값보다 오른쪽 값이 더 크다  |
| >=  | 오른쪽 값보다 왼쪽 값이 더 크거나 같다  |
| <=  | 왼쪽 값보다 오른쪽 값이 더 크거나 같다  |

```SQL
postgres=# SELECT * FROM developers WHERE book_id = 1;
 book_id |    date    |  name
---------+------------+--------
       1 | 2019-12-17 | "자바"
(1개 행)
```

- `=` 대신에 다른 비교 연산자를 입력하도록 합니다. 
```SQL
postgres=# SELECT * FROM developers WHERE book_id <> 3;
 book_id |    date    |       name
---------+------------+------------------
       1 | 2019-12-17 | "자바"
       2 | 2019-12-17 | '자바'
       4 | 2019-12-18 | 맛있는 MongoDB
       5 | 2019-12-18 | Python
       6 | 2019-12-18 | R
       7 | 2019-12-19 | Ruby
       8 | 2019-12-19 | Go's Programming
       9 | 2019-12-19 | Statistics
(8개 행)
```

- WHERE 조건에는 between과 OR 연산자를 추가할 수 있습니다. 
```SQL
postgres=# SELECT * FROM developers WHERE book_id between 3 and 5;
 book_id |    date    |         name
---------+------------+----------------------
       3 | 2019-12-17 | Python's Programming
       4 | 2019-12-18 | 맛있는 MongoDB
       5 | 2019-12-18 | Python
(3개 행)
```

- 이번에는 OR절을 같이 활용하도록 합니다. 
```SQL
postgres=# SELECT * FROM developers WHERE book_id between 2 and 3 OR name = '
Python';
 book_id |    date    |         name
---------+------------+----------------------
       2 | 2019-12-17 | '자바'
       3 | 2019-12-17 | Python's Programming
       5 | 2019-12-18 | Python
(3개 행)
```
